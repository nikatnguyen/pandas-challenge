```python
# % passing math (above 60% is passing)
passedmath =  len(students_df[students_df['math_score']>70])


#Using total students in district from earlier to calculate %
percentpass_math = passedmath/totalstudentsindistrict * 100
print(percentpass_math)
```

```python
# % passing reading
passedread = len(students_df[students_df['reading_score']>70])

#Using total students in district from earlier to calculate %
percentpass_reading = passedread/totalstudentsindistrict * 100
print(percentpass_reading)
```
